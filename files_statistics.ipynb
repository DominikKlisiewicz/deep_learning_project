```python
import os
from pydub import AudioSegment
data_dir_path = os.path.join(os.getcwd(), 'data')
```

```python
def get_audio_duration(path):
    audio = AudioSegment.from_file(path, format="flac")
    duration_ms = len(audio)
    duration_sec = duration_ms / 1000.0
    return duration_sec
```

```python
import os

data_dir_path = os.path.join(os.getcwd(), 'data', 'dev-clean')

def get_dirs(dir):
    entries = os.listdir(dir)
    directories = [entry for entry in entries if os.path.isdir(os.path.join(dir, entry))]
    return directories

readers = get_dirs(data_dir_path)

n = 0
audio_lengths = {}
files_number_per_reader = {}

print("Directories and flac files in the data directory:")
for reader_directory in readers:
    print(reader_directory)
    files_number_per_reader[reader_directory] = 0
    reader_path = os.path.join(data_dir_path, reader_directory)
    dirs_layer_1 = get_dirs(reader_path)
    for dir1 in dirs_layer_1:
        print(f"-{dir1}")
        dir_path2 = os.path.join(reader_path, dir1)
        entries = os.listdir(dir_path2)
        flac_files = [entry for entry in entries if os.path.splitext(os.path.join(dir_path2, entry))[1] == ".flac"]
        number_of_files = len(flac_files)
        n += number_of_files
        files_number_per_reader[reader_directory] += number_of_files
        for file in flac_files:
            print(f"--{file}")
            path = os.path.join(dir_path2, file)
            duration = get_audio_duration(path)
            audio_lengths[file] = duration
            print(f"{duration} seconds")
print(n)
print(files_number_per_reader)
print(audio_lengths)
```

```python
import pandas as pd
data = [{'File': k, 'Length': v} for k, v in audio_lengths.items()]
df = pd.DataFrame(data)
```

```python
df['reader'] = df['File'].str.split('-').str[0]
filter_ = (df["Length"] > 4) & (df["Length"] <6)
df[filter_]
```
